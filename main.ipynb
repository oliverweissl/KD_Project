```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors

%matplotlib inline
dpi = 100 #dpi -> controls plot size
```

```python
#for querying ncbi taxonomy dataset
from ete3 import NCBITaxa
ncbi = NCBITaxa()
```

```python
#get data from CSV
df = pd.read_csv('codon_usage.csv',low_memory=False)
dfs = np.split(df, [5], axis=1)

codons = dfs[1]
species = dfs[0]

codons = codons[sorted(codons.columns)]
pd.set_option("display.max_rows", 10, "display.max_columns", None)
```

Visialize Codon for Species

```python
def get_codon_distr(X): #Get distribution of Codons
    X = np.array(X).astype(np.float)
    first_ACGU = [sum(X[:16]), sum(X[16:32]),sum(X[32:48]), sum(X[48:64])]
    second_ACGU = []
    for i in range(4):
        comp = sum(X[i*16:i*16+4]) + sum(X[i*16+4:i*16+8]) + sum(X[i*16+8:i*16+12]) +sum(X[i*16+12: i*16+16])
        
        second_ACGU.append(sum(X[i*16:i*16+4])/comp*first_ACGU[i])
        second_ACGU.append(sum(X[i*16+4:i*16+8])/comp*first_ACGU[i])
        second_ACGU.append(sum(X[i*16+8:i*16+12])/comp*first_ACGU[i])
        second_ACGU.append(sum(X[i*16+12: i*16+16])/comp*first_ACGU[i])
    return first_ACGU, second_ACGU, X


def get_codon_graph(index): #Plot codon distribution
    organism, organism_name = codons.loc[index], species.loc[index][4]
    vals = get_codon_distr(organism)

    fig, ax = plt.subplots(dpi=dpi) 
    
    radius = 1
    size = radius/3
    
    cmap = plt.get_cmap("tab20c")
    inner_colors = cmap([1,5,10,17])

    #rings of the plot
    inner, _ = ax.pie(vals[2], radius= radius, colors = inner_colors)
    middle, _ = ax.pie(vals[1], radius= radius-size, colors = inner_colors)
    outer, _ = ax.pie(vals[0], radius = radius-2*size, colors = inner_colors)

    #legend and title
    ax.set(aspect="equal", title='Codon Usage\n'+organism_name+'')
    ax.legend(['Adenine', "Cytosine", "Guanine", "Uracil"])
    
    plt.setp(inner, edgecolor='white')
    plt.setp(middle, edgecolor='white')
    plt.setp(outer, edgecolor='white')
    
    plt.figure(figsize=(10,10))
    plt.show()

 
#get_codon_graph(13008)
```

compare two organisms

```python
def compare(o1, o2):
    org1, org2 = np.array(codons.loc[o1]).astype(np.float), np.array(codons.loc[o2]).astype(np.float)
    o1_name, o2_name = species.loc[o1][4],species.loc[o2][4]
    
    similarity = []
    for i in range(len(org1)):
        similarity.append((org1[i]/org2[i]) if org2[i] > org1[i] else (org2[i]/org1[i]))
    overall_sim = sum(similarity)/len(similarity)
    
    
    n_groups = 64
    index = np.arange(n_groups)
    bar_width = 0.4
    opacity = 0.8
    
    
    fig, ax = plt.subplots(dpi=dpi)
    plt.rcParams["figure.autolayout"] = True
    plt.rcParams["figure.figsize"] = (12,4)
    
   
    ax1 = fig.add_axes(ax.get_position(), sharex=ax, sharey=ax)
    ax1.set_facecolor('none')
    ax1.set_axis_off()
    ax2 = ax.twinx()
    ax2.set_ylim([0, 1])
    ax.tick_params(axis='x', rotation=90)

        
    sim, = ax2.plot(similarity, color='orange')
    rects1 = ax1.bar(index, org1, bar_width, alpha=opacity,color='b')
    rects2 = ax1.bar(index+bar_width, org2, bar_width, alpha=opacity,color='r')
    
    
    plt.xlabel('Codon')
    plt.ylabel('Percentage')
    plt.title('Codon Percentage for Organism')
    plt.xticks(index + bar_width, codons.columns)
    ax.legend([rects1, rects2, sim], [o1_name, o2_name, 'Similarity'],bbox_to_anchor=(1.05, 0.1), loc='upper left', borderaxespad=0)
    
    plt.show()
    return overall_sim

compare(13008, 360)
```
